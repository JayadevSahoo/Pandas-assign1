# Using a loop, calculate the average gdpPercap for all the countries and store them in an array.

```python
coun=df.country.unique()
```

```python
l = []
for i in df.country.unique():
    l.append(df['gdpPercap'][df.country == i].mean())
```

```python
avg_gdp = np.array(l)
```
